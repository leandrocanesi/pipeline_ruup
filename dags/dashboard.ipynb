```python
from airflow import DAG
from airflow.operators.empty import EmptyOperator
from airflow.providers.papermill.operators.papermill import PapermillOperator
from airflow.datasets import Dataset
from datetime import datetime

gold_dataset = Dataset("/home/kenote_ubuntu/projetos/Airflow/data/gold/forecast_summary")

with DAG(
    dag_id="generate_dashboard_on_gold_update",
    description="ðŸ“Š Gera dashboard com dados consolidados apÃ³s update na camada Gold",
    schedule=[gold_dataset],  # Gatilho via Dataset
    start_date=datetime(2025, 4, 15),
    catchup=False,
    tags=["dashboard", "visualization"],
) as dag:

    start = EmptyOperator(task_id="start")

    run_dashboard = PapermillOperator(
        task_id="generate_dashboard",
        input_nb="/home/kenote_ubuntu/projetos/Airflow/notebooks/05_dashboard.ipynb",
        output_nb="/home/kenote_ubuntu/projetos/Airflow/notebooks/05_dashboard_out.ipynb"
    )

    success_marker = EmptyOperator(task_id="success_marker")

    start >> run_dashboard >> success_marker

```
